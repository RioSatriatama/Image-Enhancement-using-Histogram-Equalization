# **Case Study 2: Image Enhancement using Histogram Equalization**

**Objective: To improve the contrast of an image by implementing histogram equalization.**

Intruction :

1. Load an Image: Read an image (preferably a low-contrast image).
2. Convert to Grayscale: Convert the image to grayscale to perform histogram equalization.
3. Perform Histogram Equalization: Apply histogram equalization to enhance the image contrast.
4. Visualize the Results: Display the original and enhanced images side by side, along with their histograms.

Expected Output:

* The original and contrast-enhanced images.
* Histograms showing the pixel intensity distribution before and after equalization.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load an image
image = cv2.imread('/content/Controll Room.jpeg', 0)

# Step 2: Perform histogram equalization
equalized_image = cv2.equalizeHist(image)

# Step 3: Display original and equalized images
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')

# Step 4: Plot histograms of the original and equalized images
plt.subplot(2, 2, 3)
plt.hist(image.ravel(), bins=256, range=[0, 256])
plt.title('Original Histogram')

plt.subplot(2, 2, 4)
plt.hist(equalized_image.ravel(), bins=256, range=[0, 256])
plt.title('Equalized Histogram')

plt.tight_layout()
plt.show()
```

# **Conclusion for Case Study 2: Image Enhancement using Histogram Equalization**

In this case study, we successfully enhanced the contrast of a low-contrast image using histogram equalization. The original image displayed a narrow range of pixel intensity values, resulting in poor contrast. By applying histogram equalization, we redistributed the pixel intensities to cover a broader range, which improved the overall contrast of the image.

This process is particularly effective for images where the details are obscured due to poor lighting or other conditions that lead to limited dynamic range. The equalized image showed more visible details and balanced contrast compared to the original. The histograms further confirmed this improvement, with the equalized image displaying a more uniform distribution of pixel intensities, spanning a wider range.

Thus, histogram equalization proves to be a powerful technique for enhancing image quality in scenarios where contrast is a limiting factor.
